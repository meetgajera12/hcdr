```python
df['AGE'] = (-df['DAYS_BIRTH'] / 365).astype(int)

df['YEARS_EMPLOYED'] = round(abs(df['DAYS_EMPLOYED']) / 365, 2)

df['YEARS_CURRENT_PHONE'] = round(abs(df['DAYS_LAST_PHONE_CHANGE']) / 365, 2)
```

```python
df.drop(
    columns=[
        'DAYS_BIRTH',
        'DAYS_EMPLOYED',
        'DAYS_LAST_PHONE_CHANGE'
    ],
    inplace=True
)
```

```python
df['CRDIT_INCOME_RATIO'] = round(df['AMT_CREDIT']/df['AMT_INCOME_TOTAL'], 4)
```

```python
df['ANNUITY_INCOME_RATIO'] = round(df['AMT_ANNUITY']/df['AMT_INCOME_TOTAL'], 4)
```

```python
df['CREDIT_ANNUITY_RATIO'] = round(df['AMT_CREDIT']/df['AMT_ANNUITY'], 4)
```

```python
df['GOODS_CREDIT_RATIO'] = round(df['AMT_GOODS_PRICE']/df['AMT_CREDIT'], 4)
```

```python
df['INCOME_PER_FAM_MEMBER'] = round(df['AMT_INCOME_TOTAL']/df['CNT_FAM_MEMBERS'], 4)
```

```python
df['EMPLOYMENT_AGE_RATIO'] = round(df['YEARS_EMPLOYED']/df['AGE'], 4)
```

```python
df['IS_SINGLE_PARENT'] = (
                          (df['CNT_CHILDREN']>0) 
                          & 
                          (df['NAME_FAMILY_STATUS'].isin(['Married','Civil marriage']))
                          ).astype(int)
```
