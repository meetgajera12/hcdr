## Post-Merge Missing Value Handling & Feature Engineering

```python
# bureau history indicator

merge_df1['HAS_BUREAU_HISTORY'] = (
    merge_df1['AMT_CREDIT_SUM_sum'].notna().astype(int)
)

merge_df1.insert(
    merge_df1.columns.get_loc('AMT_CREDIT_SUM_sum'),
    'HAS_BUREAU_HISTORY',
    merge_df1.pop('HAS_BUREAU_HISTORY')
)
```

```python
# Count features

count_cols = [
    'HAS_DEBT_sum',
    'IS_ACTIVE_sum',
    'IS_CLOSED_sum',
    'IS_SOLD_sum',
    'IS_BAD_DEBT_sum',
    'AMT_CREDIT_SUM_DEBT_MISSING_sum',
    'AMT_CREDIT_SUM_LIMIT_MISSING_sum'
]

merge_df1[count_cols] = merge_df1[count_cols].fillna(0)
```

```python
# Amount features

amount_cols = [
    'AMT_CREDIT_SUM_sum',
    'AMT_CREDIT_SUM_mean',
    'AMT_CREDIT_SUM_max',
    'AMT_CREDIT_SUM_DEBT_sum',
    'AMT_CREDIT_SUM_DEBT_mean',
    'AMT_CREDIT_SUM_LIMIT_sum'
]

merge_df1[amount_cols] = merge_df1[amount_cols].fillna(0)
```

```python
#  Ratio features

ratio_cols = [
    'DEBT_RATIO_mean',
    'DEBT_RATIO_max',
    'CREDIT_UTILIZATION_mean'
]

merge_df1[ratio_cols] = merge_df1[ratio_cols].fillna(0)
```

```python
# Time features

time_cols = [
    'YEARS_CREDIT_min',
    'YEARS_CREDIT_mean',
    'YEARS_CREDIT_max'
]

for col in time_cols:
    merge_df1[col] = merge_df1[col].fillna(merge_df1[col].median())
```

`YEARS_CREDIT` features represent the age of a customer's credit history. Filling missing values with `0` would incorrectly imply that the customer obtained credit today (0 years ago). Therefore, missing values were imputed using the median to preserve a realistic representation of credit age.

```python
# missing indicators

merge_df1[
    ['AMT_CREDIT_SUM_DEBT_MISSING_mean',
        'AMT_CREDIT_SUM_LIMIT_MISSING_mean']
] = merge_df1[
    ['AMT_CREDIT_SUM_DEBT_MISSING_mean',
        'AMT_CREDIT_SUM_LIMIT_MISSING_mean']
].fillna(0)
```
