```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df0 = pd.read_csv(r'C:\Users\meet.gajera\OneDrive\Desktop\hcdr__\Data\row\application_train.csv')
```

```python
col = ['SK_ID_CURR','TARGET',
       'CODE_GENDER','DAYS_BIRTH','CNT_CHILDREN','CNT_FAM_MEMBERS',
       'NAME_CONTRACT_TYPE','AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE',
       'NAME_INCOME_TYPE','OCCUPATION_TYPE','ORGANIZATION_TYPE','DAYS_EMPLOYED',
       'NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS',
       'FLAG_OWN_CAR','FLAG_OWN_REALTY','NAME_HOUSING_TYPE',
       'DAYS_LAST_PHONE_CHANGE','FLAG_EMAIL','FLAG_WORK_PHONE']

df = df0[col].copy()

df = pd.DataFrame(data=df,columns=col)
```

### CNT_FAM_MEMBERS

```python
print('mean:',df['CNT_FAM_MEMBERS'].mean())
print('median',df['CNT_FAM_MEMBERS'].median())
```

```python
# filling NA by Median
df['CNT_FAM_MEMBERS'] = df['CNT_FAM_MEMBERS'].fillna(
    df['CNT_FAM_MEMBERS'].median()
)
```

### DAYS_LAST_PHONE_CHANGE

```python
print('mean:',df['DAYS_LAST_PHONE_CHANGE'].mean())
print('median',df['DAYS_LAST_PHONE_CHANGE'].median())
```

```python
# filling NA by Median
df['DAYS_LAST_PHONE_CHANGE'] = df['DAYS_LAST_PHONE_CHANGE'].fillna(
    df['DAYS_LAST_PHONE_CHANGE'].median()
)
```

### AMT_ANNUITY

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
df['AMT_ANNUITY'].plot(kind='kde')
plt.subplot(1,2,2)
df['AMT_ANNUITY'].plot(kind='box')
plt.show()

print("skewness:", df['AMT_ANNUITY'].skew())
```

```python
# filling NA
df['AMT_ANNUITY'].fillna(df['AMT_ANNUITY'].median(), inplace=True)
```

```python
# transformation
df['AMT_ANNUITY_trans'] = np.log1p(df['AMT_ANNUITY'])
```

```python
df.insert(
    df.columns.get_loc('AMT_ANNUITY') + 1,
    'AMT_ANNUITY_trans',
    df.pop('AMT_ANNUITY_trans')
)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
df['AMT_ANNUITY_trans'].plot(kind='kde')
plt.subplot(1,2,2)
df['AMT_ANNUITY_trans'].plot(kind='box')
plt.show()
```

```python
print("Before:", df['AMT_ANNUITY'].skew())
print("After :", df['AMT_ANNUITY_trans'].skew())
```

### AMT_GOODS_PRICE

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
df['AMT_GOODS_PRICE'].plot(kind='kde')
plt.subplot(1,2,2)
df['AMT_GOODS_PRICE'].plot(kind='box')
plt.show()

print('skewness: ',df['AMT_GOODS_PRICE'].skew())
```

```python
# filling NA
df['AMT_GOODS_PRICE'].fillna(df['AMT_GOODS_PRICE'].median(), inplace=True)
```

```python
df['AMT_GOODS_PRICE'].skew()    # after filling missing values by median
```

```python
# transformation
df['AMT_GOODS_PRICE_trans'] = np.log1p(df['AMT_GOODS_PRICE'])

df.insert(
    df.columns.get_loc('AMT_GOODS_PRICE') + 1,
    'AMT_GOODS_PRICE_trans',
    df.pop('AMT_GOODS_PRICE_trans')
)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
df['AMT_GOODS_PRICE_trans'].plot(kind='kde')
plt.subplot(1,2,2)
df['AMT_GOODS_PRICE_trans'].plot(kind='box')
plt.show()
```

### OCCUPATION_TYPE

```python
df['OCCUPATION_TYPE'].value_counts()
```

```python
df['OCCUPATION_TYPE'].value_counts().plot(kind='bar')
```

```python
# missing indicator
df['OCCUPATION_TYPE_MISSING'] = df['OCCUPATION_TYPE'].isna().astype(int)  

df.insert(
    df.columns.get_loc('OCCUPATION_TYPE')+1,
    'OCCUPATION_TYPE_MISSING',
    df.pop('OCCUPATION_TYPE_MISSING')
)

# filling NA
df['OCCUPATION_TYPE'] = df['OCCUPATION_TYPE'].fillna('Unknown')
```

```python
df['OCCUPATION_TYPE_MISSING'].value_counts()
```

```python
df.isnull().sum()
```

```python
# drop all transformed columns.
df.drop(columns=df.filter(like='trans').columns, inplace=True)
```
